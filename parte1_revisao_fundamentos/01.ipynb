```
import math 

def numprimo(num):
    '''
     Verifica se um valor é primo
    '''
    if (num %2) == 0 and num > 2:
        return "Este numero não é primo"
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if (num % i) == 0:
            return "Este numero não é primo"
    else:
        return "Numero primo."
```

```
numprimo(int(input("Numero = ")))
```
