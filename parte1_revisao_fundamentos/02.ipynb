###Dividindo em colunas

```
s = open('salarios.csv', 'r')
data = s.read()
rows = data.split('\n')
full_data = []

for row in rows:
    split_row = row.split(",")
    full_data.append(split_row)
```

```
print(full_data)
```

###Contando as linhas de um arquivo

```
count = 0
for row in full_data:
    count +=1
print(count)
```
